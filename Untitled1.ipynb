# Exploratory Data Analysis of FIFA World Cup Data

The FIFA World Cup is one of the most prestigious and widely watched sporting events globally, featuring national teams competing for the title of world champion in men's football (soccer). In this project, we'll perform exploratory data analysis (EDA) on FIFA World Cup data to uncover interesting insights and patterns.

```python
import os
import pandas as pd
```

```python
data_dir = 'extracted_data'
```

```python
world_cup_matches_df = pd.read_csv(os.path.join(data_dir, 'WorldCupMatches.csv'))
world_cup_players_df = pd.read_csv(os.path.join(data_dir, 'WorldCupPlayers.csv'))
world_cups_df = pd.read_csv(os.path.join(data_dir, 'WorldCups.csv'))
```

```python
print("World Cup Matches DataFrame:")
print(world_cup_matches_df.info())

print("\nWorld Cup Players DataFrame:")
print(world_cup_players_df.info())

print("\nWorld Cups DataFrame:")
print(world_cups_df.info())
```

```python
# Dropping missing values
world_cup_matches_df.dropna(inplace=True)
world_cup_players_df['Position'].fillna(world_cup_players_df['Position'].mode()[0], inplace=True)
world_cup_players_df['Event'].fillna('No event', inplace=True)
```

```python
# Summary statistics
print("\nSummary statistics for World Cups DataFrame:")
print(world_cups_df.describe())

print("\nSummary statistics for World Cup Matches DataFrame:")
print(world_cup_matches_df.describe())

print("\nSummary statistics for World Cup Players DataFrame:")
print(world_cup_players_df.describe())
```

```python
import matplotlib.pyplot as plt

home_goals = world_cup_matches_df.groupby('Home Team Name')['Home Team Goals'].sum()
away_goals = world_cup_matches_df.groupby('Away Team Name')['Away Team Goals'].sum()
total_goals_per_team = home_goals.add(away_goals, fill_value=0)
total_goals_per_team = total_goals_per_team.sort_values(ascending=False)

plt.figure(figsize=(12, 8))
total_goals_per_team.plot(kind='bar', color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Total Goals Scored by Each Team in World Cup Matches')
plt.xlabel('Team')
plt.ylabel('Total Goals Scored')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
goals_by_stage = world_cup_matches_df.groupby('Stage').agg({'Home Team Goals': 'sum', 'Away Team Goals': 'sum', 'MatchID': 'nunique'})
goals_by_stage['Total Goals'] = goals_by_stage['Home Team Goals'] + goals_by_stage['Away Team Goals']
goals_by_stage['Average Goals per Match'] = goals_by_stage['Total Goals'] / goals_by_stage['MatchID']

plt.figure(figsize=(12, 6))
goals_by_stage[['Home Team Goals', 'Away Team Goals']].plot(kind='bar', stacked=True)
plt.title('Goals Scored by Stage')
plt.xlabel('Stage')
plt.ylabel('Goals')
plt.xticks(rotation=45)
plt.legend(['Home Team Goals', 'Away Team Goals'])
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.barplot(x=total_goals_per_team.values, y=total_goals_per_team.index, palette='viridis')
plt.title('Total Goals Scored by Each Team')
plt.xlabel('Total Goals Scored')
plt.ylabel('Team')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(y=world_cup_players_df['Position'], order=world_cup_players_df['Position'].value_counts().index, palette='viridis')
plt.title('Distribution of Player Positions')
plt.xlabel('Count')
plt.ylabel('Position')
plt.show()
```

```python
players_goals = world_cup_players_df.groupby('Player Name')['Event'].apply(lambda x: x.str.count('G').sum()).sort_values(ascending=False)
top_scorers = players_goals.head(10)
print("Top Scorers of All Time:")
print(top_scorers)
```
